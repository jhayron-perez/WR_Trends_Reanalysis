```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import xarray as xr
import glob
import xesmf as xe
import os

def regrid(ds, variable, reuse_weights=False,filename_weights = None):
    """
    Function to regrid onto coarser ERA5 grid (0.25-degree).
    Args:
        ds (xarray dataset): file.
        variable (str): variable.
        reuse_weights (boolean): Whether to use precomputed weights to speed up calculation.
                                 Defaults to ``False``.
        filename_weights (str): if reuse_weights is True, then a string for the weights path is needed.
    Returns:
        Regridded data file for use with machine learning model.
    """
    # ds.lon = ds.longitude
    # ds.lat = ds.latitude
    # ds = ds.rename({'latitude':'lat',
    #                'longitude':'lon'})
    ds_out = xe.util.grid_2d(lon0_b=0-0.5,   lon1_b=360-0.5, d_lon=1., 
                             lat0_b=-90-0.5, lat1_b=90,      d_lat=1.)
    
    if reuse_weights == False:
        regridder = xe.Regridder(ds, ds_out, method='nearest_s2d', reuse_weights=reuse_weights)
        return regridder(ds[variable]),regridder
    else:
        regridder = xe.Regridder(ds, ds_out, method='nearest_s2d', reuse_weights=reuse_weights,
            filename = filename_weights)
        return regridder(ds[variable])
```

```python
import os
import re
from collections import defaultdict

# Define the folder path
folder_path = '/glade/derecho/scratch/jhayron/Z500_JRA3Q_Daily/'

# Initialize a dictionary to count files per year
file_count_per_year = defaultdict(int)

# Define a regular expression to extract the year from the file names
file_pattern = re.compile(r'jraq_z500_(\d{4})_')

# Loop through the files in the folder
for file_name in os.listdir(folder_path):
    match = file_pattern.search(file_name)
    if match:
        year = match.group(1)
        file_count_per_year[year] += 1

# Print the result
for year, count in sorted(file_count_per_year.items()):
    print(f"Year {year}: {count} files")
```

```python
import os
import xarray as xr

# Function to process NetCDF files in a given directory
def process_files_in_directory(directory, variable):
    files = [f for f in os.listdir(directory) if f.endswith('.nc')]
    data_list = []

    for file in files:
        file_path = os.path.join(directory, file)
        ds = xr.open_dataset(file_path)
        if variable in ds.variables:
            data_list.append(ds[variable].values)
        ds.close()
    
    return data_list

# Base path to the directories containing the NetCDF files
base_path = '/glade/campaign/collections/rda/data/d640000/anl_p/'

# Loop through each year and month
for year in range(1947, 2024):
    for month in range(1, 13):
        directory = os.path.join(base_path, f'{year}{month:02d}')
        if os.path.exists(directory):
            print(f'Processing directory: {directory}')
            data = process_files_in_directory(directory, 'hgt')
            # Perform any further processing on the extracted data
            # For example, concatenate, average, etc.
            print(f'Extracted data for {year}-{month:02d}')
```
